### Descripción del notebook:

Este notebook tiene la finalidad de crear un conjunto de datos para el proyecto.


Para las url del tipo 'phishing' se ha utilizado PhishTank, la URL para acceder es el siguiente: https://phishtank.org/developer_info.php


Para las url legítimas o benignas se ha utilizado del portal Mendeley Data. En este conjunto contiene tantas url benignas y maliciosas, pero para el proyecto solo nos interesa las urls benignas por lo que se realizará un filtrado. La URL es el siguiente:
https://data.mendeley.com/datasets/gdx3pkwp47/2

Por último para las url del tipo 'malware' y 'defacement', se han utilizado del portal de Kaggle. Igualmente en este conjunto contiene url de otros tipos pero se ha realizado un filtrado para obtener sólo las url del tipo mencionadas, la URL es el siguiente: \
https://www.kaggle.com/datasets/sid321axn/malicious-urls-dataset

```python
import pandas as pd
```

```python
import warnings
warnings.filterwarnings('ignore')
```

Subconjunto de URL del tipo phishing

```python
phishing = 'http://data.phishtank.com/data/online-valid.csv.gz'

phish = pd.read_csv(phishing)
```

```python
print(f"El conjunto phishing tiene {len(phish)} url")
```

Subconjunto de url del tipo benigno

```python
benign = pd.read_csv('Webpages_Classification_train_data.csv')
```

```python
benign['label'].value_counts()
```

```python
print("Este conjunto tiene 1.172.747 url del tipo benigno")
```

Subconjunto de url de los tipos defacement y malware

```python
malicious = pd.read_csv('malicious_phish-kaggle.csv')
```

```python
malicious['type'].value_counts()
```

```python
print("Este conjunto tiene aproximadamente 96000 y 32000 url del tipo defacement y malware respectivamente")
```

### Filtrado de los subconjuntos

Filtrar y quedarse solamente con la columna 'url' de cada subconjunto

```python
phish_url = phish[['url']]
```

```python
phish_url
```

```python
benign_url = benign[benign['label'] == 'good']
```

```python
benign_url['label'].unique()
```

```python
print("Asegurar que unicamente hay url benignas")
```

```python
benign_url = benign_url[['url']]
```

```python
benign_url
```

```python
malware_url = malicious[malicious['type'] == 'malware']
```

```python
malware_url
```

```python
malware_url = malware_url[['url']]
```

```python
malware_url
```

```python
deface_url = malicious[malicious['type'] == 'defacement']
```

```python
deface_url
```

```python
deface_url = deface_url[['url']]
```

```python
deface_url
```

### Creación de una nueva columna

Añadir una nueva columna del tipo de url 'type' a cada uno de los subconjuntos

```python
benign_url['type'] = 'benign'
benign_url
```

```python
malware_url['type'] = 'malware'
malware_url
```

```python
deface_url['type'] = 'defacement'
deface_url
```

```python
phish_url['type'] = 'phishing'
phish_url
```

### Extracción de muestras

Extraer 5000 muestras de cada subconjunto con la idea de crear un conjunto de datos balanceado

```python
benign_url = benign_url.sample(n = 5000, random_state=23)
benign_url = benign_url.reset_index(drop=True)
```

```python
deface_url = deface_url.sample(n = 5000, random_state=23)
deface_url = deface_url.reset_index(drop=True)
```

```python
malware_url = malware_url.sample(n = 5000, random_state=23)
malware_url = malware_url.reset_index(drop=True)
```

```python
phish_url = phish_url.sample(n = 5000, random_state=23)
phish_url = phish_url.reset_index(drop=True)
```

```python
benign_url
```

```python
malware_url
```

```python
phish_url
```

```python
deface_url
```

### Creación del conjunto de datos

Unir los subconjuntos en el conjunto de datos final que se utilizará en el proyecto

```python
dataset_final = pd.concat([benign_url, malware_url, phish_url, deface_url]).reset_index(drop=True)
```

```python
dataset_final
```

```python
dataset_final['type'].value_counts()
```

Mezclar las filas del dataset final

```python
dataset_final = dataset_final.sample(frac=1, random_state=23)
```

Finalmente guardar el dataset creado

```python
dataset_final.to_csv('dataset_final.csv',index=False)
```
